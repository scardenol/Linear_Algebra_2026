```python
# Forma escalonada reducida
# Factorización rango
# Inversa generalizada

import numpy as np
from sympy import Matrix

# Definir la matriz
B = Matrix([[1, -2, 4, 3], [0, 1, 2, -1], [2, -5, 6, 7]])

# 1. Calcular la forma escalonada reducida (rref)
# rref() devuelve una tupla: (matriz_reducida, columnas_pivote)

BR, pivotes = B.rref()

#Imprimir los resultados hasta aquí para validar que sea lo esperado

print("Matriz original:")
print("\nB = ")
display(B)
print("\nMatriz reducida (RREF):")
print("\nBR = ")
display(BR)
print("\npivotes =",pivotes)
```

```python
#2. Factorización rango

# Crear una nueva matriz F con las columnas pivote de la matriz original
# Crear una nueva matriz G con las filas pivote de la matriz reducida
# La sintaxis : indica todas las filas o las columnas

p = len(pivotes)
print("\nNúmero de pivotes=", p)

F = B[:, [j for j in pivotes]]
G = BR[0:p, :]

# Verificar que B=FG es una factorización rango de la matriz original

VFactor=F@G-B #El resultado debe ser la matriz nula

print("\nMatrices de la factorización:")
print("\nF = ")
display(F)
print("\nG = ")
display(G)
print("\nVFactor = ")
display(VFactor)
```

```python
#3. Creación de una inversa generalizada de la matriz original B
print("\nInversa generalizada:")

#Inversa generalizada de F, denominada Finvg
F1 = F.T@F
F1 = F1.inv()
Finvg = F1@F.T
print("\nFinvg = ")
display(Finvg)

#Verificar que Finvg es inversa a izquierda de F
I1=Finvg@F
print("\n(Finvg)F =") #Debe imprimir la matriz identidad
display(I1)

#Inversa generalizada de G, denominada Ginvg
G1 = G@G.T
G1 = G1.inv()
Ginvg = G.T@G1
print("\nGinvg = ")
display(Ginvg)

#Verificar que Ginvg es inversa a derecha de G
I2 = G@Ginvg
print("\nG(Ginvg) = ") #Debe imprimir la matriz identidad
display(I2)

#Inversa generalizada de B, denominada Binvg
Binvg = Ginvg@Finvg

print("\nBinvg = ")
display(Binvg)

#Verificar que la matriz Binvg satisface la igualdad B=B(Binvg)B

B2 = B@Binvg@B

print("\nB(Binvg)B = ")
display(B2)
```
