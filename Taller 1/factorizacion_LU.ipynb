```python
# Factorización LU con SymPy
# Devuelve las matrices L, U, y los indices de las filas permutadas P de la matriz M
from sympy import Matrix
B = Matrix([[2, -1, 0, 0], [6, -4, 5, -3], [8, -4, 1, 0], [4, -1, 0, 7]])
L, U, P = B.LUdecomposition()
```
